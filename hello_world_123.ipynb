```python
aadhar_no = input("Enter your 12-digit Aadhaar number: ")

if len(aadhar_no) == 12 and aadhar_no.isdigit():
    print("Aadhaar verified successfully.")
    income_str = input("Enter your monthly income: ")
    credit_score_str = input("Enter your credit score (300-850): ")
    debt_str = input("Enter your total existing monthly debt payments: ")

    income = float(income_str)
    credit_score = int(credit_score_str)
    debt = float(debt_str)
    min_income = 2000.00
    min_credit_score = 650
    max_debt_to_income_ratio = 0.35
    is_eligible = True
    reasons_for_ineligibility = []

    print("\n--- Checking Eligibility ---")
    if income < min_income:
        is_eligible = False
        reasons_for_ineligibility.append(f"Income (${income:.2f}) is below minimum.")

    if credit_score < min_credit_score:
        is_eligible = False
        reasons_for_ineligibility.append(f"Credit score ({credit_score}) is below minimum.")

    debt_to_income_ratio = debt / income if income > 0 else float('inf')
    if debt_to_income_ratio > max_debt_to_income_ratio:
        is_eligible = False
        reasons_for_ineligibility.append(f"Debt-to-income ratio ({debt_to_income_ratio:.2%}) is too high.")

    if is_eligible:
        print("Congratulations! You are eligible for a loan.")
    else:
        print("Not eligible for a loan. Reasons:")
        for reason in reasons_for_ineligibility:
            print(f"- {reason}")
    print("\n--- Eligibility Check Complete ---")
else:
    print("Invalid Aadhaar number. It must be exactly 12 digits.")
```

```python
account_holders = {"C101", "C102", "C103", "C104", "C105", "C106"}
kyc_verified = {"C102", "C104", "C106", "C108"}
defaulters = {"C107", "C105"}

# 1️⃣ Create set of customers eligible for loan
eligible_customers = account_holders & kyc_verified - defaulters

print("Eligible customers:", eligible_customers)

# 2️⃣ Display total eligible customers
print("Total eligible customers:", len(eligible_customers))

# 3️⃣ Check whether C104 is eligible
print("Is C104 eligible?", "C104" in eligible_customers)

# 4️⃣ Customers either verified or defaulters but not both
result = kyc_verified ^ defaulters
print("Verified or Defaulters but not both:", result)

# 5️⃣ Update operations
kyc_verified.add("C107") # add C107
account_holders.discard("C105") # remove safely
print("Updated KYC verified:", kyc_verified)
print("Updated account holders:", account_holders)
# 6️⃣ Clean up
backup = kyc_verified.copy()
kyc_verified.clear()
print("Backup KYC:", backup)
print("Original KYC after clear:", kyc_verified)
```

```python

daily_transactions = [5000, 1500, -200, 12000, 300, -500, 700000]

print("All Transactions:")
print(daily_transactions)

# Step 2: Filter Deposits (only positive values)
deposits = list(filter(lambda x: x > 0, daily_transactions))

print("\nFiltered Deposits (Positive Transactions):")
print(deposits)

# Step 3: Map for Commission (0.05% commission)
commission_rates = list(map(lambda x: x * 0.0005, deposits))

print("\nCommission on each deposit (0.05%):")
print(commission_rates)

# Step 4: Reduce to calculate total commission
total_commission = reduce(lambda a, b: a + b, commission_rates)

print("\nTotal Commission Earned:")
print(total_commission)
```

```python
flights = {
    "AI202": 10,
    "6E501": 5
}

passengers = []

def show_status():
    print("\nCurrent Seat Availability")
    for name, count in flights.items():
        print("Flight", name, ":", count, "seats left")

    print("\nPassenger List")
    for p in passengers:
        print(p)

def book_ticket():
    name = input("Enter Passenger Name: ")
    age = input("Enter Age: ")
    phone = input("Enter Phone Number: ")
    f_id = input("Enter Flight ID (AI202 or 6E501): ")

    if f_id in flights:
        if flights[f_id] > 0:
            flights[f_id] = flights[f_id] - 1

            passenger = {
                "name": name,
                "age": age,
                "phone": phone,
                "flight": f_id
            }

            passengers.append(passenger)
            print("Ticket booked successfully for", name)
        else:
            print("No seats available")
    else:
        print("Flight not found")

def cancel_ticket():
    name = input("Enter Passenger Name to cancel: ")

    for p in passengers:
        if p["name"] == name:
            flights[p["flight"]] = flights[p["flight"]] + 1
            passengers.remove(p)
            print("Ticket cancelled for", name)
            return

    print("Passenger not found")

def add_seats():
    f_id = input("Enter Flight ID: ")
    seats = int(input("Enter number of seats to add: "))

    if f_id in flights:
        flights[f_id] = flights[f_id] + seats
        print(seats, "seats added to", f_id)
    else:
        print("Flight not found")

def generate_ticket():
    name = input("Enter Passenger Name: ")

    for p in passengers:
        if p["name"] == name:
            print("\n----- TICKET -----")
            print("Passenger Name:", p["name"])
            print("Age:", p["age"])
            print("Phone:", p["phone"])
            print("Flight ID:", p["flight"])
            print("------------------")
            return

    print("Passenger not found")

while True:
    print("\nAIRLINE SYSTEM")
    print("1 View Flights")
    print("2 Book Ticket")
    print("3 Cancel Ticket")
    print("4 Exit")
    print("5 Add Seat Availability")
    print("6 Generate Ticket")

    choice = input("Enter choice: ")

    if choice == "1":
        show_status()
    elif choice == "2":
        book_ticket()
    elif choice == "3":
        cancel_ticket()
    elif choice == "5":
        add_seats()
    elif choice == "6":
        generate_ticket()
    elif choice == "4":
        print("System Closed")
        break
    else:
        print("Invalid choice")
```
